1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sympy as sp
```

```python
# Визначення функції та її похідної
x = np.linspace(0, 10)
f = 10 * x - 5
df = np.full_like(x, 10) # Похідна f'(x) = 10

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x, f, label='f(x) = 10x - 5', color='blue')
plt.plot(x, df, label="f'(x) = 10", color='green')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Визначаємо функції
f_sin = sp.sin(x)
f_ln = sp.ln(x)

# Обчислюємо похідні
f_sin_df = sp.diff(f_sin, x)
f_ln_df = sp.diff(f_ln, x)

# Створюємо діапазон значень x
x_vals_sin = np.linspace(-2*np.pi, 2*np.pi, 400) #беремо діапазон від -2pi до 2pi
x_vals_ln = np.linspace(0.1, 10, 400)  # для ln(x) починаємо з 0.1 щоб уникнути ln(0)

# Обчислюємо значення функцій та їх похідних
y_sin = [f_sin.evalf(subs={x: val}) for val in x_vals_sin]
y_sin_df = [f_sin_df.evalf(subs={x: val}) for val in x_vals_sin]
y_ln = [f_ln.evalf(subs={x: val}) for val in x_vals_ln]
y_ln_df = [f_ln_df.evalf(subs={x: val}) for val in x_vals_ln]

# Створюємо графіки
plt.figure(figsize=(12, 6))

# Графік функції sin(x) та її похідної
plt.subplot(1, 2, 1)
plt.plot(x_vals_sin, y_sin, label='f(x) = sin(x)', color='blue')
plt.plot(x_vals_sin, y_sin_df, label="f'(x) = cos(x)", color='green')
plt.title('Функція sin(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

# Графік функції ln(x) та її похідної
plt.subplot(1, 2, 2)
plt.plot(x_vals_ln, y_ln, label='f(x) = ln(x)', color='blue')
plt.plot(x_vals_ln, y_ln_df, label="f'(x) = 1/x", color='green')
plt.title('Функція ln(x) та її похідна')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
# Визначаємо символічну змінну
x = sp.symbols('x')

# Квадратична функція
f = x**2 - 5*x + 1
derivative_f = sp.diff(f, x)

print("Похідна функції x^2 -5x + 1:", derivative_f)

# Створюємо діапазон значень x
x_vals = np.linspace(0, 100, 5000) #беремо діапазон від 0 до 100

# Обчислюємо значення функції та їх похідної
y = [f.evalf(subs={x: val}) for val in x_vals]
y_derivative = [derivative_f.evalf(subs={x: val}) for val in x_vals]

# Створення графіку
plt.figure(figsize=(10, 5))
plt.plot(x_vals, y, label='f(x) = x^2 -5x + 1', color='blue')
plt.plot(x_vals, y_derivative, label="f'(x)", color='green')
plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```
